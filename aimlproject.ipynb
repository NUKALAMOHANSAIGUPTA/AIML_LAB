```python
# Required Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Load your dataset
data = pd.read_csv("/content/diabetes.csv")  # Adjust as needed

# Step 2: Preprocess the dataset (Assuming your target is in a column named 'Outcome')
X = data.drop(columns=['Outcome'])  # Features
y = data['Outcome']  # Target

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train the SVM model
svm_model = SVC(kernel='linear')  # You can choose different kernels like 'rbf', 'poly', etc.
svm_model.fit(X_train, y_train)

# Step 5: Make predictions
y_pred = svm_model.predict(X_test)

# Step 6: Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
error_rate = 1 - accuracy
conf_matrix = confusion_matrix(y_test, y_pred)

# Step 7: Print metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"Error Rate: {error_rate:.2f}")
print(f"Confusion Matrix:\n{conf_matrix}")

# Step 8: Plot all metrics
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot Accuracy
axs[0, 0].bar(['Accuracy'], [accuracy], color='blue')
axs[0, 0].set_ylim(0, 1)
axs[0, 0].set_title("Accuracy")
axs[0, 0].set_ylabel("Score")

# Plot Error Rate
axs[0, 1].bar(['Error Rate'], [error_rate], color='red')
axs[0, 1].set_ylim(0, 1)
axs[0, 1].set_title("Error Rate")
axs[0, 1].set_ylabel("Score")

# Plot Recall
axs[1, 0].bar(['Recall'], [recall], color='purple')
axs[1, 0].set_ylim(0, 1)
axs[1, 0].set_title("Recall")
axs[1, 0].set_ylabel("Score")

# Plot Precision
axs[1, 1].bar(['Precision'], [precision], color='cyan')
axs[1, 1].set_ylim(0, 1)
axs[1, 1].set_title("Precision")
axs[1, 1].set_ylabel("Score")

# Adjust layout
plt.tight_layout()

# Step 9: Show plots
plt.show()

# Step 10: Plot the Confusion Matrix
ConfusionMatrixDisplay(conf_matrix).plot()
plt.title("Confusion Matrix")
plt.show()
```

```python
# Required Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Set the plotting theme
plt.style.use('seaborn')  # Change the theme here if you want

# Step 2: Load your dataset
data = pd.read_csv("/content/diabetes.csv") # Adjust as needed

# Step 3: Preprocess the dataset (Assuming your target is in a column named 'Outcome')
X = data.drop(columns=['Outcome'])  # Features
y = data['Outcome']  # Target

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Step 6: Make predictions
y_train_pred = dt_model.predict(X_train)
y_test_pred = dt_model.predict(X_test)

# Step 7: Calculate metrics for both training and testing sets
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)

train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)

train_error_rate = 1 - train_accuracy
test_error_rate = 1 - test_accuracy

train_conf_matrix = confusion_matrix(y_train, y_train_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)

# Step 8: Plot all metrics
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot Accuracy
axs[0, 0].bar(['Train Accuracy', 'Test Accuracy'], [train_accuracy, test_accuracy], color=['blue', 'green'])
axs[0, 0].set_ylim(0, 1)
axs[0, 0].set_title("Train vs Test Accuracy")
axs[0, 0].set_ylabel("Accuracy")

# Plot Error Rate
axs[0, 1].bar(['Train Error Rate', 'Test Error Rate'], [train_error_rate, test_error_rate], color=['red', 'orange'])
axs[0, 1].set_ylim(0, 1)
axs[0, 1].set_title("Train vs Test Error Rate")
axs[0, 1].set_ylabel("Error Rate")

# Plot Recall
axs[1, 0].bar(['Train Recall', 'Test Recall'], [train_recall, test_recall], color=['purple', 'yellow'])
axs[1, 0].set_ylim(0, 1)
axs[1, 0].set_title("Train vs Test Recall")
axs[1, 0].set_ylabel("Recall")

# Plot Precision
axs[1, 1].bar(['Train Precision', 'Test Precision'], [train_precision, test_precision], color=['cyan', 'magenta'])
axs[1, 1].set_ylim(0, 1)
axs[1, 1].set_title("Train vs Test Precision")
axs[1, 1].set_ylabel("Precision")

# Adjust layout
plt.tight_layout()

# Step 9: Show plots
plt.show()

# Step 10: Plot the Confusion Matrix for both Training and Testing sets
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Confusion matrix for Training data
ConfusionMatrixDisplay(train_conf_matrix).plot(ax=ax[0], colorbar=False)
ax[0].set_title("Confusion Matrix - Training")

# Confusion matrix for Testing data
ConfusionMatrixDisplay(test_conf_matrix).plot(ax=ax[1], colorbar=False)
ax[1].set_title("Confusion Matrix - Testing")

# Show Confusion Matrices
plt.show()
print(f"Train Recall: {train_recall}")
print(f"Test Recall: {test_recall}")
print(f"Precision: {precision}")
print(f"Accuracy: {accuracy}")
print(f"Error Rate: {error_rate}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
# Required Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Step 1: Set the plotting theme
plt.style.use('seaborn')  # Change the theme here if you want

# Step 2: Load your dataset
data = pd.read_csv("/content/diabetes.csv")  # Adjust as needed

# Step 3: Preprocess the dataset (Assuming your target is in a column named 'Outcome')
# Dropping some columns (for example, 'Pregnancies' and 'SkinThickness')
X = data.drop(columns=['Outcome', 'Pregnancies', 'SkinThickness'])  # Features
y = data['Outcome']  # Target

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train the Linear Regression model
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)

# Step 6: Make predictions
y_train_pred = lin_reg_model.predict(X_train)
y_test_pred = lin_reg_model.predict(X_test)

# Step 7: Convert predictions to binary outcomes using a threshold of 0.5
y_train_pred_binary = (y_train_pred >= 0.5).astype(int)
y_test_pred_binary = (y_test_pred >= 0.5).astype(int)

# Step 8: Calculate metrics for both training and testing sets
train_accuracy = accuracy_score(y_train, y_train_pred_binary)
test_accuracy = accuracy_score(y_test, y_test_pred_binary)

train_recall = recall_score(y_train, y_train_pred_binary)
test_recall = recall_score(y_test, y_test_pred_binary)

train_precision = precision_score(y_train, y_train_pred_binary)
test_precision = precision_score(y_test, y_test_pred_binary)

train_error_rate = 1 - train_accuracy
test_error_rate = 1 - test_accuracy

train_conf_matrix = confusion_matrix(y_train, y_train_pred_binary)
test_conf_matrix = confusion_matrix(y_test, y_test_pred_binary)

# Step 9: Print all metrics
print("Training Metrics:")
print(f"  Accuracy: {train_accuracy:.2f}")
print(f"  Recall: {train_recall:.2f}")
print(f"  Precision: {train_precision:.2f}")
print(f"  Error Rate: {train_error_rate:.2f}")
print(f"  Confusion Matrix:\n{train_conf_matrix}")

print("\nTesting Metrics:")
print(f"  Accuracy: {test_accuracy:.2f}")
print(f"  Recall: {test_recall:.2f}")
print(f"  Precision: {test_precision:.2f}")
print(f"  Error Rate: {test_error_rate:.2f}")
print(f"  Confusion Matrix:\n{test_conf_matrix}")

# Step 10: Plot all metrics
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot Accuracy
axs[0, 0].bar(['Train Accuracy', 'Test Accuracy'], [train_accuracy, test_accuracy], color=['blue', 'green'])
axs[0, 0].set_ylim(0, 1)
axs[0, 0].set_title("Train vs Test Accuracy")
axs[0, 0].set_ylabel("Accuracy")

# Plot Error Rate
axs[0, 1].bar(['Train Error Rate', 'Test Error Rate'], [train_error_rate, test_error_rate], color=['red', 'orange'])
axs[0, 1].set_ylim(0, 1)
axs[0, 1].set_title("Train vs Test Error Rate")
axs[0, 1].set_ylabel("Error Rate")

# Plot Recall
axs[1, 0].bar(['Train Recall', 'Test Recall'], [train_recall, test_recall], color=['purple', 'yellow'])
axs[1, 0].set_ylim(0, 1)
axs[1, 0].set_title("Train vs Test Recall")
axs[1, 0].set_ylabel("Recall")

# Plot Precision
axs[1, 1].bar(['Train Precision', 'Test Precision'], [train_precision, test_precision], color=['cyan', 'magenta'])
axs[1, 1].set_ylim(0, 1)
axs[1, 1].set_title("Train vs Test Precision")
axs[1, 1].set_ylabel("Precision")

# Adjust layout
plt.tight_layout()

# Step 11: Show plots
plt.show()

# Step 12: Plot the Confusion Matrix for both Training and Testing sets
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Confusion matrix for Training data
ConfusionMatrixDisplay(train_conf_matrix).plot(ax=ax[0], colorbar=False)
ax[0].set_title("Confusion Matrix - Training")

# Confusion matrix for Testing data
ConfusionMatrixDisplay(test_conf_matrix).plot(ax=ax[1], colorbar=False)
ax[1].set_title("Confusion Matrix - Testing")

# Show Confusion Matrices
plt.show()
```
